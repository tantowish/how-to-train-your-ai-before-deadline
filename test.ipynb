```python
# convert the date_time column to a datetime datatype
df['DateTime'] = pd.to_datetime(df['DateTime'])
```

```python
# drop Zone 2 and Zone 3 columns, as we are only interested in Zone 1
df = df.drop(columns=['Zone 2  Power Consumption', 'Zone 3  Power Consumption'])

df.head()
```

```python
df.head()
```

```python
df['Day'] = df['DateTime'].dt.day
df['Month'] = df['DateTime'].dt.month
df['Hour'] = df['DateTime'].dt.hour
df['Minute'] = df['DateTime'].dt.minute
df['Day of Week'] = df['DateTime'].dt.dayofweek + 1
```

```python
df.head()
```

```python
def month_to_quarter(series):
    if series['Month'] <=3:
        return 1
    elif series['Month'] >3 and series['Month'] <=6:
        return 2
    elif series['Month'] >6 and series['Month'] <=9:
        return 3
    elif series['Month'] >9 and series['Month'] <=12:
        return 4
```

```python
df['Quarter of Year'] = df.apply(month_to_quarter, axis='columns')
df['Day of Year'] = df['DateTime'].dt.strftime('%j').astype(int)  
df = df.drop(['DateTime'],axis=1)
```

```python
df.columns
```

```python
df = df[['Temperature', 'Humidity', 'Wind Speed', 'general diffuse flows',
       'diffuse flows', 'Day', 'Month', 'Hour',
       'Minute', 'Day of Week', 'Quarter of Year', 'Day of Year','Zone 1 Power Consumption']]
```

```python
df.head()
```

```python
corr = df.corr().round(decimals=3)
mask = np.triu(np.ones_like(corr, dtype=np.bool))
f, ax = plt.subplots(figsize=(20, 20))
cmap = sns.diverging_palette(240, 10, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=None, center=0,square=True, annot=True, linewidths=.5, cbar_kws={"shrink": .9})
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

zone1 = sns.boxplot(data=df, x='Hour', y='Zone 1 Power Consumption', palette = 'Blues', showfliers=False)

plt.suptitle('KW by Hour', fontsize=15)
plt.xlabel('hour', fontsize=12)
plt.ylabel('Power Consumption in KW', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

sns.boxplot(data=df, x='Hour', y='Temperature', palette = 'Reds', showfliers=False)

plt.suptitle('Temperature by Hour', fontsize=15)
plt.xlabel('hour', fontsize=12)
plt.ylabel('Temperature in Â°C', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```

```python

fig, ax = plt.subplots(figsize=(20, 10))

sns.boxplot(data=df, x='Hour', y='Humidity', palette = 'Greens', showfliers=False)

plt.suptitle('Humidity by Hour', fontsize=15)
plt.xlabel('hour', fontsize=12)
plt.ylabel('Humidity in %', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```
