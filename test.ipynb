```python
# convert the date_time column to a datetime datatype
df['DateTime'] = pd.to_datetime(df['DateTime'])
```

```python
# drop Zone 2 and Zone 3 columns, as we are only interested in Zone 1
df = df.drop(columns=['Zone 2  Power Consumption', 'Zone 3  Power Consumption'])

df.head()
```

```python
df.head()
```

```python
df['Day'] = df['DateTime'].dt.day
df['Month'] = df['DateTime'].dt.month
df['Hour'] = df['DateTime'].dt.hour
df['Minute'] = df['DateTime'].dt.minute
df['Day of Week'] = df['DateTime'].dt.dayofweek + 1
```

```python
df.head()
```

```python
def month_to_quarter(series):
    if series['Month'] <=3:
        return 1
    elif series['Month'] >3 and series['Month'] <=6:
        return 2
    elif series['Month'] >6 and series['Month'] <=9:
        return 3
    elif series['Month'] >9 and series['Month'] <=12:
        return 4
```

```python
df['Quarter of Year'] = df.apply(month_to_quarter, axis='columns')
df['Day of Year'] = df['DateTime'].dt.strftime('%j').astype(int)  
df = df.drop(['DateTime'],axis=1)
```

```python
df.columns
```

```python
df = df[['Temperature', 'Humidity', 'Wind Speed', 'general diffuse flows',
       'diffuse flows', 'Day', 'Month', 'Hour',
       'Minute', 'Day of Week', 'Quarter of Year', 'Day of Year','Zone 1 Power Consumption']]
```

## Split Train and Test Data

```python
train_size = int(len(df) * 0.8)
train_df = df[:train_size]
test_df = df[train_size:]

# Visualize the split
fig, ax = plt.subplots(figsize=(25, 8))

# Plotting power consumption to visualize split
ax.plot(train_df.index, train_df['Zone 1 Power Consumption'], label='Training Set', color='darkorange')
ax.plot(test_df.index, test_df['Zone 1 Power Consumption'], label='Test Set', color='yellowgreen')

# Mark the point of split
ax.axvline(train_size, color='black', linestyle='--')

# Add legend, labels, and title
ax.legend(['Training Set', 'Test Set'])
plt.title('Data Train/Test Split', fontsize=15)
plt.xlabel('Index', fontsize=12)
plt.ylabel('Power Consumption in KW', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Display the plot
plt.show()
```
